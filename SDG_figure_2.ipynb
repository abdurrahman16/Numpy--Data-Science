```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(12, 7))

# Buildings
for x in [1, 3, 5]:
    ax.add_patch(patches.Rectangle((x, 3), 1.5, 1.5, fill=True, color='lightgray'))
    ax.text(x+0.75, 3.75, "Building", ha='center')

# Solar panels
for x in [1.1, 3.1, 5.1]:
    ax.add_patch(patches.Rectangle((x, 2.5), 1.3, 0.3, color='gold'))
ax.text(3.75, 2.3, "Solar Panels", ha='center')

# Battery storage
ax.add_patch(patches.Rectangle((8, 3), 1.5, 1.5, fill=True, color='skyblue'))
ax.text(8.75, 3.75, "Battery\nStorage", ha='center')

# Control center
ax.add_patch(patches.Rectangle((8, 1), 1.5, 1.5, fill=True, color='lightgreen'))
ax.text(8.75, 1.75, "Control\nCenter", ha='center')

# Lines from buildings to battery
for x in [2, 4, 6]:
    ax.plot([x, 8], [3.75, 3.75], 'k-')

# Grid-connected / island mode switch
ax.add_patch(patches.Rectangle((4, 0.5), 2, 0.5, fill=False))
ax.text(5, 0.75, "Grid â†” Island Mode", ha='center')

# Remove axes
ax.set_xlim(0, 11)
ax.set_ylim(0, 5)
ax.axis('off')

plt.title("Local Microgrid Model", fontsize=18)
plt.show()
```
